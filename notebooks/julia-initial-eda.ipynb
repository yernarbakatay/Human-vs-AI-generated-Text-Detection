```python
import pandas as pd
import plotly.figure_factory as ff
import plotly.express as px

df = pd.read_csv("data/transformed_dataset.csv")

df.head()
```

```python
df["ai_target"].value_counts()
```

```python
df.info()
```

```python
df['response_length'] = df['response'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df.head()
```

```python
# plot the distribution of the length of the responses split by ai_target
fig = px.histogram(
    df,
    x='response_length',
    color='ai_target',
    nbins=30,
    barmode='group',
    title='Distribution of Response Lengths by AI Target',
    text_auto=True,
    color_discrete_sequence=px.colors.qualitative.Set2
 )
fig.update_traces(marker_line_color='black', marker_line_width=1)
fig.update_layout(
    xaxis_title='Number of Words in Response',
    yaxis_title='Frequency',
    legend_title='AI Target',
    template='plotly_white'
 )
fig.show()
```

```python

# separate density plot of response length by ai_target
target_values = list(df['ai_target'].dropna().unique())
hist_data = [
    df.loc[df['ai_target'] == target, 'response_length'].dropna().tolist()
    for target in target_values
 ]
group_labels = [str(target) for target in target_values]
colors = px.colors.qualitative.Set2[:len(group_labels)]

fig_density = ff.create_distplot(
    hist_data,
    group_labels,
    show_hist=False,
    show_rug=False,
    colors=colors
 )
fig_density.update_layout(
    title='Density of Response Lengths by AI Target',
    xaxis_title='Number of Words in Response',
    yaxis_title='Density',
    legend_title='AI Target',
    template='plotly_white'
 )
fig_density.show()
```

```python
# plot for responses >2000
df_long_responses = df[df['response_length'] > 2000]
fig_long = px.histogram(
    df_long_responses,
    x='response_length',
    color='ai_target',
    nbins=30,
    barmode='group',
    title='Distribution of Long Response Lengths (>2k) by AI Target',
    text_auto=True,
    color_discrete_sequence=px.colors.qualitative.Set2
 )
fig_long.update_traces(marker_line_color='black', marker_line_width=1)
fig_long.update_layout(
    xaxis_title='Number of Words in Response',
    yaxis_title='Frequency',
    legend_title='AI Target',
    template='plotly_white'
 )
fig_long.show()
```

```python
# plot for responses <=2000
df_short_responses = df[df['response_length'] <= 2000]
fig_long = px.histogram(
    df_short_responses,
    x='response_length',
    color='ai_target',
    nbins=30,
    barmode='group',
    title='Distribution of Short Response Lengths (<2k) by AI Target',
    text_auto=True,
    color_discrete_sequence=px.colors.qualitative.Set2
 )
fig_long.update_traces(marker_line_color='black', marker_line_width=1)
fig_long.update_layout(
    xaxis_title='Number of Words in Response',
    yaxis_title='Frequency',
    legend_title='AI Target',
    template='plotly_white'
 )
fig_long.show()
```
