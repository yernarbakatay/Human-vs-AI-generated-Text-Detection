```python
from datasets import load_dataset
import pandas as pd
dataset = load_dataset(
    "csv",
    data_files="https://huggingface.co/datasets/gsingh1-py/train/resolve/main/train.csv"
)

df = dataset["train"].to_pandas()
df
```

#Initial Word Count

```python
columns_to_word_count = [
    'Human_story',
    'gemma-2-9b',
    'mistral-7B',
    'qwen-2-72B',
    'llama-8B',
    'accounts/yi-01-ai/models/yi-large',
    'GPT_4-o'
]

for col in columns_to_word_count:
    if col in df.columns:
        df[col] = df[col].astype(str)
        df[f'{col}_word_count'] = df[col].apply(lambda x: len(str(x).split()))
    else:
        print(f"Column '{col}' not found in DataFrame.")

display(df.head(20))
```

```python
df.info()
```

#Dataset Reframe

```python
columns_to_process = [
    'Human_story',
    'gemma-2-9b',
    'mistral-7B',
    'qwen-2-72B',
    'llama-8B',
    'accounts/yi-01-ai/models/yi-large',
    'GPT_4-o'
]

reshaped_data = []

for col in columns_to_process:
    if col in df.columns and f'{col}_word_count' in df.columns:
        temp_df = df[['prompt', col, f'{col}_word_count']].copy()

        temp_df.columns = ['prompt', 'text', 'word_count']

        temp_df['ai_target'] = 0 if col == 'Human_story' else 1

        temp_df['source_model'] = col

        reshaped_data.append(temp_df)

new_df = pd.concat(reshaped_data, ignore_index=True)

print(f"Original rows: {len(df)}")
print(f"New rows: {len(new_df)} (Expected: {len(df) * 7})")

display(new_df.head(30))
```

```python
new_df.info()
```

#Word Count 2 for Data Check

```python
new_df['text'] = new_df['text'].astype(str)
new_df['word_count2'] = new_df['text'].apply(lambda x: len(str(x).split()))

new_df['is_match'] = new_df['word_count'] == new_df['word_count2']
mismatch_count = len(new_df[new_df['is_match'] == False])

print(f"Number of mismatched records: {mismatch_count}")
if mismatch_count == 0:
    print("Word counts for all records match! No data loss.")
else:
    print("Word counts for some records do not match. Please check the data.")

display(new_df.head())
```

#Adding text100 and word count for text 100 (just in case)

```python
# Drop records where ai_target is 1 and word count is less than 200
initial_len = len(new_df)
new_df = new_df[~((new_df['ai_target'] == 1) & (new_df['word_count'] < 200))].copy()
dropped_count = initial_len - len(new_df)

print(f"Dropped {dropped_count} records where ai_target=1 and word count is less than 200.")
print(f"Remaining records: {len(new_df)}")

# Function to extract the required 100 words based on ai_target
def extract_text100(row):
    words = str(row['text']).split()
    if row['ai_target'] == 0:
        # First 100 words (if less than 100, keeps all)
        return ' '.join(words[:100])
    else:
        # Skip first 100 words, take the next 100 words
        return ' '.join(words[100:200])

# Apply the function to create the text100 column
new_df['text100'] = new_df.apply(extract_text100, axis=1)

# Calculate the word count of the new text100 column to verify
new_df['text100_word_count'] = new_df['text100'].apply(lambda x: len(str(x).split()))

# Display the results to verify
display(new_df[['ai_target', 'word_count', 'text100_word_count', 'text100']].head(10))
```

#Adding text200 and word count for text 200 (just in case)

```python
# Drop records where ai_target is 1 and word count is less than 300
initial_len = len(new_df)
new_df = new_df[~((new_df['ai_target'] == 1) & (new_df['word_count'] < 300))].copy()
dropped_count = initial_len - len(new_df)

print(f"Dropped {dropped_count} records where ai_target=1 and word count is less than 300.")
print(f"Remaining records: {len(new_df)}")

# Function to extract the required 200 words based on ai_target
def extract_text200(row):
    words = str(row['text']).split()
    if row['ai_target'] == 0:
        # First 200 words (if less than 200, keeps all)
        return ' '.join(words[:200])
    else:
        # Skip first 100 words, take the next 200 words (words 101 to 300)
        return ' '.join(words[100:300])

# Apply the function to create the text200 column
new_df['text200'] = new_df.apply(extract_text200, axis=1)

# Calculate the word count of the new text200 column to verify
new_df['text200_word_count'] = new_df['text200'].apply(lambda x: len(str(x).split()))

# Display the results to verify
display(new_df[['ai_target', 'word_count', 'text200_word_count', 'text200']].head(10))
```

#Edit: Adding txt30 and 50


```python
# Function to extract the required 30 words based on ai_target
def extract_text30(row):
    words = str(row['text']).split()
    if row['ai_target'] == 0:
        # First 30 words
        return ' '.join(words[:30])
    else:
        # Skip first 100 words, take the next 30 words
        return ' '.join(words[100:130])

# Function to extract the required 50 words based on ai_target
def extract_text50(row):
    words = str(row['text']).split()
    if row['ai_target'] == 0:
        # First 50 words
        return ' '.join(words[:50])
    else:
        # Skip first 100 words, take the next 50 words
        return ' '.join(words[100:150])

# Apply the functions
new_df['text30'] = new_df.apply(extract_text30, axis=1)
new_df['text30_word_count'] = new_df['text30'].apply(lambda x: len(str(x).split()))

new_df['text50'] = new_df.apply(extract_text50, axis=1)
new_df['text50_word_count'] = new_df['text50'].apply(lambda x: len(str(x).split()))

display(new_df[['ai_target', 'text30_word_count', 'text50_word_count', 'text30', 'text50']].head(10))
```
